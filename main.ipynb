Pandas Profiling is a Python library that allows us to generate a very detailed report on our pandas dataframe without much input from the user.It passes 3 steps:
1) Summarizing dataset
2) Generating report structure
3) Render HTML.
For saving time on Exploratory Data Analysis (EDA),This library is really helpful.

Let's start.

```python
#At fisrt we should import pandas
import pandas as pd
#Don't forget to install pandas_profiling
#!pip install -U pandas-profiling[notebook]
#Now it's time to import PandasProfiling
import pandas_profiling
```

```python
#Define dataset
DataSet1=pd.read_csv("bank_marketing_dataset.csv")
```

```python
#Now let 's see what will happen!
pandas_profiling.ProfileReport(DataSet1)
```
